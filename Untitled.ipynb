```python
import os
import numpy as np
import cv2
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing import image
```

```python
# 1. Load and preprocess images
def load_data(data_dir, img_size=224):
    classes = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
    X, y = [], []

    for label in classes:
        path = os.path.join(data_dir, label)
        for img_file in os.listdir(path):
            img_path = os.path.join(path, img_file)
            img = cv2.imread(img_path, cv2.IMREAD_COLOR)
            if img is not None:
                img = cv2.resize(img, (img_size, img_size))
                X.append(img)
                y.append(classes.index(label))  # Convert label to an index

    X = np.array(X) / 255.0  # Normalize images
    y = np.array(y)
    return X, y
```

```python
# 2. Load training and testing data
train_dir = "E:\\college\\grad project\\integrated\\training"  # Set your training dataset directory
test_dir = "E:\\college\\grad project\\integrated\\testing"    # Set your testing dataset directory

X_train, y_train = load_data(train_dir)  
X_test, y_test = load_data(test_dir)
```

```python
# 3. Encode labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# 4. Define CNN model
def create_cnn_model(input_shape=(224, 224, 3)):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(4, activation='softmax')  # 4 classes
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

cnn_model = create_cnn_model()
```

```python
# 5. Load ResNet50V2 pre-trained model
resnet = ResNet50V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = resnet.output
x = GlobalAveragePooling2D()(x)
output = Dense(4, activation='softmax')(x)  # 4 classes
resnet_model = Model(inputs=resnet.input, outputs=output)
resnet_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# 6. Data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
datagen.fit(X_train)
```

```python
# 7. Train CNN and ResNet50V2 models
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
# Train CNN
cnn_model.fit(datagen.flow(X_train, y_train_encoded, batch_size=32), 
              validation_data=(X_test, y_test_encoded), 
              epochs=20, 
              callbacks=[early_stopping])
```

```python
# Train ResNet50V2
resnet_model.fit(datagen.flow(X_train, y_train_encoded, batch_size=16), 
                 validation_data=(X_test, y_test_encoded), 
                 epochs=20, 
                 callbacks=[early_stopping])
```

```python
# 8 Predict features in batches
def predict_in_batches(model, data, batch_size=32):
    predictions = []
    for i in range(0, len(data), batch_size):
        batch = data[i:i+batch_size]
        batch_predictions = model.predict(batch)
        predictions.append(batch_predictions)
    return np.concatenate(predictions, axis=0)

cnn_features_train = predict_in_batches(cnn_model, X_train)
cnn_features_test = predict_in_batches(cnn_model, X_test)

resnet_features_train = predict_in_batches(resnet_model, X_train)
resnet_features_test = predict_in_batches(resnet_model, X_test)
```

```python
# 9. Flatten original images for stacking
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)
```

```python
# 10. Function to combine features using memory mapping
def save_combined_features(file_name, features_list):
    shapes = [f.shape for f in features_list]
    total_shape = (features_list[0].shape[0], sum([s[1] for s in shapes]))  # Calculate final shape

    # Create memory-mapped file to write combined data
    combined_features = np.memmap(file_name, dtype='float32', mode='w+', shape=total_shape)
    
    # Fill the memory-mapped file incrementally
    start_col = 0
    for feature in features_list:
        end_col = start_col + feature.shape[1]
        combined_features[:, start_col:end_col] = feature
        start_col = end_col
    
    combined_features.flush()  # Ensure all data is written to disk
    return combined_features
```
