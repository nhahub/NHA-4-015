```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications import ResNet50
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import load_model
from sklearn.model_selection import train_test_split
```

```python
# Set GPU memory growth to avoid memory issues
physical_devices = tf.config.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
# Define paths to your training and test datasets
train_dir = "E:\\college\\grad project\\integrated\\training"  # replace with the actual path to the training data
test_dir = "E:\\college\\grad project\\integrated\\testing"  # replace with path to test data
```

```python
# Parameters
batch_size = 32
img_size = (224, 224)  # ResNet50 input size
num_classes = 4  # Number of cancer types
```

```python
# Data augmentation and generators for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # Split 20% of the data for validation
)

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python

# Create train and validation generators by splitting the data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'  # Use the training subset of the split data
)

val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'  # Use the validation subset of the split data
)
```

```python
# Test data generator (no splitting, as it's your held-out test set)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # Important to preserve the order for evaluation
)
```

```python
# Build the model using ResNet50 as base
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Freeze the base model layers to prevent retraining during the initial training
base_model.trainable = False
```

```python
# Define the custom top layers for your model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1024, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Set callbacks for early stopping and model checkpointing
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)
```
