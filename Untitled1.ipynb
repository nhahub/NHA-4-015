```python
def brain_cancer_risk_assessment():
    print("Answer the following questions about symptoms, lab results, and pathological indicators.")
    
    # Symptoms
    persistent_headache = input("Do you have persistent headaches? (yes/no): ").lower() == "yes"
    nausea_vomiting = input("Do you experience frequent nausea or vomiting? (yes/no): ").lower() == "yes"
    seizures = input("Have you experienced any seizures? (yes/no): ").lower() == "yes"
    cognitive_changes = input("Do you have memory problems or confusion? (yes/no): ").lower() == "yes"
    vision_problems = input("Do you have visual disturbances (e.g., blurry or double vision)? (yes/no): ").lower() == "yes"

    # Risk factors
    family_history = input("Do you have a family history of brain tumors? (yes/no): ").lower() == "yes"
    prior_radiation = input("Have you undergone radiation therapy to the head? (yes/no): ").lower() == "yes"
    immune_suppression = input("Do you have immune suppression (e.g., HIV, transplant)? (yes/no): ").lower() == "yes"
    genetic_condition = input("Do you have a genetic condition like NF1, NF2, or Li-Fraumeni syndrome? (yes/no): ").lower() == "yes"

    # Pathological indicators
    abnormal_mri = input("Has your MRI indicated a tumor with concerning features? (yes/no): ").lower() == "yes"
    biopsy_confirmed = input("Has a biopsy confirmed malignancy? (yes/no): ").lower() == "yes"
    idh_mutation = input("Does your tumor have an IDH mutation? (yes/no): ").lower() == "yes"
    mgmt_methylation = input("Is MGMT promoter methylation present? (yes/no): ").lower() == "yes"
    elevated_ldh = input("Do you have elevated lactate dehydrogenase (LDH)? (yes/no): ").lower() == "yes"
    csf_tumor_markers = input("Are tumor markers or malignant cells found in CSF? (yes/no): ").lower() == "yes"
    ki_67_high = input("Is the Ki-67 proliferation index high? (yes/no): ").lower() == "yes"

    # Adjusted weights for all indicators
    weights = {
        # Clinical symptoms
        "persistent_headache": 0.1,
        "nausea_vomiting": 0.05,
        "seizures": 0.15,
        "cognitive_changes": 0.1,
        "vision_problems": 0.1,
        # Risk factors
        "family_history": 0.15,
        "prior_radiation": 0.2,
        "immune_suppression": 0.1,
        "genetic_condition": 0.25,
        # Pathological indicators
        "abnormal_mri": 0.4,
        "biopsy_confirmed": 0.5,
        "idh_mutation": -0.1,  # Protective factor
        "mgmt_methylation": -0.05,  # Protective factor
        "elevated_ldh": 0.2,
        "csf_tumor_markers": 0.3,
        "ki_67_high": 0.3,
    }
    
    # Calculate total risk score
    score = 0
    for factor, weight in weights.items():
        response = locals()[factor]
        score += weight if response else 0

    # Normalize score to a probability percentage
    max_score = sum(w for w in weights.values() if w > 0)  # Exclude protective factors
    probability = (score / max_score) * 100

    # Risk categorization
    if probability >= 80:
        risk_level = "Very High Risk"
    elif probability >= 60:
        risk_level = "High Risk"
    elif probability >= 40:
        risk_level = "Moderate Risk"
    elif probability >= 20:
        risk_level = "Low Risk"
    else:
        risk_level = "Very Low Risk"

    # Output results
    print("\n--- Brain Cancer Risk Assessment ---")
    print(f"Probability of Brain Cancer: {probability:.2f}%")
    print(f"Risk Level: {risk_level}")
    print("-----------------------------------")

# Run the program
if __name__ == "__main__":
    brain_cancer_risk_assessment()
```
