```python
import numpy as np
from tensorflow.keras.models import load_model
from PIL import Image
```

```python
# Load the models
resnet50_model = load_model("saved_model/mri_classifier.keras")
custom_cnn_model = load_model("saved_model/custom_cnn_classifier.keras")
```

```python
# Class names
class_names = ["glioma_tumor", "meningioma_tumor", "no_tumor", "pituitary_tumor"]
```

```python
# Function to predict using a single model
def predict_image(image_path, model):
    img = Image.open(image_path).resize((224, 224))
    img_array = np.expand_dims(np.array(img) / 255.0, axis=0)  # Normalize and batch
    predictions = model.predict(img_array)
    return class_names[np.argmax(predictions)]
```

```python
# Function to use ensemble prediction
def ensemble_predict(image_path, models):
    img = Image.open(image_path).resize((224, 224))
    img_array = np.expand_dims(np.array(img) / 255.0, axis=0)  # Normalize and batch

    predictions = [model.predict(img_array) for model in models]
    averaged_predictions = np.mean(predictions, axis=0)  # Average the predictions
    return class_names[np.argmax(averaged_predictions)]
```

```python
# Example usage
image_path = "E://college//grad project//integrated//training//glioma_tumor//G_4.jpg"
predicted_class = ensemble_predict(image_path, [resnet50_model, custom_cnn_model])
print(f"Predicted Class: {predicted_class}")
```
